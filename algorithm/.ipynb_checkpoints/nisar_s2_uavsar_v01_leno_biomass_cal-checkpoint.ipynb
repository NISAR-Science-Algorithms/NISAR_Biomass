## Canopy model for Radar Backscattering


The physical based data-fitted model (Saatchi and McDonald, 1997) describes scattering mechanisms of the forest with three components **(Fig. 1)** , namely direct **D** , Interactive **I** , i.e., direct-reflect, and ground **G** components.  

The **D** component consists of the scattering from trunks, branches, and leaves directly back to the sensor.  The **I** terms, also known as "double bounce" term, is the scattering that includes interaction between trees and ground.  This includes the trunk-ground, crown-ground scattering.  The **G** component is the scattering directly back from ground attenuated by the forest canopy.

![Figure 1. Scattering mechanisms of forest scattering model](data/fig01.png)

The model can be written in the form of the following equation,

$$
\begin{align}
\sigma_0(p,s,t) & = A_{p,t} W_s^{\alpha_p} (1 - \exp(-B_{p,t} W_s^{\beta_p})) \quad\quad\quad & [D: Volume] \\
               & + C_{p,t} W_s^{\gamma_p} \Gamma_{p,t} \exp(-B_{p,t} W_s^{\beta_p})  \quad\quad\quad & [I: Volume-Surface] \\
               & + S_{p,t} \exp(-B_{p,t} W_s^{\beta_p})  \quad\quad\quad & [G: Surface]
\end{align}
$$

where the subscripts $p, s, t$ represents the variations over $polarization, space, time$.

### Basic Assumptions

The simplified assumptions here include:

(1) Parameters $A_{p,t}, B_{p,t}, C_{p,t}$ change over time and different polarizations, but vary little spatially within a local window.

(2) Parameters $\alpha_{p}, \beta_{p}, \gamma_{p}$ change with different polarizations, but vary little temporally, or spatially within a local window.

(3) Parameters $\Gamma_{p,t}, S_{p,t}$, related to both surface roughness and dielectric properties, change over time and different polarizations, but vary little spatially within a local window.




## Further Simplifications

We rewrite the original equation as

$$
\begin{align}
\sigma_0(p,s,t) & = A_{p,t} W_s^{\alpha_p} (1 - \exp(-B_{p,t} W_s^{\beta_p})) \quad\quad\quad & [D] \\
               & + C_{p,t} W_s^{\gamma_p} \Gamma_{p,t} \exp(-B_{p,t} W_s^{\beta_p})  \quad\quad\quad & [I] \\
               & + S_{p,t} \exp(-B_{p,t} W_s^{\beta_p})  \quad\quad\quad & [G] \\
\Rightarrow \quad\quad\quad \quad\quad\quad  & \\
\sigma_0(p,s,t) & = A_{p,t} W_s^{\alpha_p} (1 - \exp(-B_{p,t} W_s)) \quad\quad\quad & [D] \\
               & + C'_{p,t} S_{t} W_s^{\gamma_p} \exp(-B_{p,t} W_s)  \quad\quad\quad & [I] \\
               & + D_{p} S_{t} \exp(-B_{p,t} W_s) \quad\quad\quad & [G]
\end{align}
$$

where we have 

**Assumption 1:**

$$
\begin{align}
\beta_p & = 1
\end{align}
$$

This assumption is consistent with the estimation of forward model simulations.

**Assumption 2:**

$S$ term (backscattering crosssection from soil rough surface, no dependency on vegetation) is separated into polarization-dependent term and time-dependent term,

$$
\begin{align}
S_{p,t} & = D_{p} S_{t}
\end{align}
$$


**Assumption 3:**

$C\Gamma$ term (reflectivity of the soil rough surface, and no dependency on vegetation) is rewritten as the new constant $C'$ with time-dependent $S_t$,

$$
\begin{align}
C_{p,t} \Gamma_{p,t} & = C_{p,t} K_{p} S_{t} \\
                     & = C'_{p,t} S_{t} 
\end{align}
$$




## Final Parameters for Retrieval Algorithm

Based on the above simplications, we have the following list of unknown parameters to solve:

$$ A_{p,t}, B_{p,t}, C'_{p,t}, D_{p}, \alpha_p, \gamma_p, S_t, W_s $$

For possible solutions, we need to satisfy the following condition,

$$ 3pt + 3p + t + s < pts $$ 

For example, a 2-polarization system of 3x3-window retrieval needs at least 2 multi-temporal observations to satisfy the condition above.

### Biomass retrieval Model 1. Universal approach (update W and S only)

Setup of the model.

1. Parameters $A$, $B$, $C'$, $D$, $\alpha$ and $\gamma$ (and $S_0$) are retrieved from calibration site

2. Initial values of AGB: $W_0 = (100 \times \sigma_{HV} )^2$

3. Update 1: Retrieval of $W_1$ from fixed parameters and $S_0$, with bounds of ±80% $W_0$

4. Update 2: Retrieval of $S_1$ from fixed parameters and $W_1$, with bounds of ±40% $S_0$

5. Update 3: Retrieval of $W_2$ from fixed parameters and $S_1$, with bounds of ±80% $W_1$

6. Retrieval is done by using local optimization – For each pixel, minimizing
    $$ Loss = \left[\omega_{HH} \left(f_{HH}(W) - \sigma_{HH, t} \right)\right]^2 + \left[\omega_{HV} \left(f_{HV}(W) - \sigma_{HV, t} \right)\right]^2 $$
   where $\omega_{HV}$ and $\omega_{HH}$ are the weights for HV and HH bands

7. For each scene, we have solutions of AGB: $\hat{W}_t$, the average solution is $\hat{W} = \sum_{t=1}^n \hat{W}_t$



* Input:
  * One LC map identifying valid pixels
  * One AGB map used for calibration
  * list of UAVSAR scenes
 
* Output:
  * Calibrated parameter file 
  * Model performance with incremental scenes

```python
#randomly select the training samples
def randomly_select_training_samples(mask_file, 
                                     number_of_samples = 500, 
                                     seed = None):
    
    import random
    
    if seed is not None:
        random.seed(seed)

    mask_train_file = mask_file
    da_train_mask = xr.open_rasterio(mask_train_file)

    train_indices = np.where(da_train_mask.values[0] > 0)
    train_indices_index = random.sample(range(0,len(train_indices[0])), number_of_samples)

    da_train =np.zeros_like(da_train_mask.values[0])

    da_train[[train_indices[0][train_indices_index],
              train_indices[1][train_indices_index]]] = 1

    train_file = 'output/mask_agb_100m_train_sampled.tif'
    mr.write_geotiff_with_gdalcopy(mask_file, da_train.reshape((-1,da_train.shape[0], da_train.shape[1])), train_file)
    
    return train_file, train_indices, train_indices_index
#     da_train = xr.open_rasterio(train_file)
#     da_train.plot()
```

```python
#randomly select the training samples
def randomly_select_testing_samples(mask_file, 
                                    train_indices, 
                                    train_indices_index, 
                                    number_of_samples = 500, seed = None):
    
    import random
    
    if seed is not None:
        random.seed(seed)
        
    mask_test_file = 'output/mask_agb_100m.tif'
    da_test_mask = xr.open_rasterio(mask_test_file)

    #mask out the training samples to avoid the replicas between training and testing samples
    da_test_mask.values[0][[train_indices[0][train_indices_index],
                            train_indices[1][train_indices_index]]] = 0
    #
    test_indices = np.where(da_test_mask.values[0] > 0)
    test_indices_index = random.sample(range(0,len(test_indices[0])), number_of_samples)

    da_test =np.zeros_like(da_test_mask.values[0])

    da_test[[test_indices[0][test_indices_index],
             test_indices[1][test_indices_index]]] = 1

    test_file = 'output/mask_agb_100m_test_sampled.tif'
    mr.write_geotiff_with_gdalcopy(mask_file, da_test.reshape((-1,da_test.shape[0], da_test.shape[1])), test_file)
    
    return test_file, test_indices, test_indices_index

#     da_test = xr.open_rasterio(test_file)
#     da_test.plot()
```

```python
# update the numbers

def update_numbers(sar_list, 
                   agb_file, 
                   mask_file, 
                   out_name, w1_noise=10, w2_noise=10):
    
    if not os.path.exists("output"):
        os.mkdir("output")
    out_prefix = f"output/{out_name}"
    mdl00 = mr.FieldRetrieval(mask_file, out_name=out_prefix)
    x0, y0, mask0 = mdl00.data_cleaner(
        sar_list,
        agb_file=agb_file,
        mask_file=mask_file,
        w1_noise=w1_noise,
        w2_noise=w2_noise,
    )
    # print(mask0)
    x0 = x0.reshape([y0.shape[0], -1, 2])
    x0v = x0[:, :, 0]
    x0h = x0[:, :, 1]

#     df = pd.DataFrame(
#         {
#             "AGB_measured": y0.ravel(),
#             "VH_mean": np.nanmean(x0v, axis=-1),
#             "VV_mean": np.nanmean(x0h, axis=-1),
#         }
#     )
#     df.to_csv(f"{out_prefix}_mean_measured.csv")
    
    return y0.ravel(), np.nanmean(x0v, axis=-1), np.nanmean(x0h, axis=-1), mask0
```

```python

# model calibration
def model_calibration(W1, mean_hv, mean_hh, init_weights=[4, 1], seed = None):
    '''
    model calibration
    '''
    mean_x = np.stack([mean_hv, mean_hh]).T
    # print(mean_x.shape)
    model_00 = mr.Retrieval(2, 1, mean_x.shape[0])
    if len(W1.shape) == 1:
        W1 = W1[:, None]
    W = model_00.x_S(W1.T)
    y = mean_x.reshape([1, -1])
    bounds = (
        (
            (0.0001, 0.5),
            (0.0001, 0.5),
        )
        + 2 * ((0.0001, 0.9999),)
        + 2 * ((0.0001, 0.9999),)
        + 2 * ((0.0001, 1.9999),)
        + 2 * ((0.0001, 3.9999),)
        + (
            (1, 1.0001),
            (0, 25),
        )
        + 1 * ((0.0001, 0.5),)
    )
    print(y.shape)
    print(W.shape)
    params, y_hat, cost = model_00.model_inverse_03a_v1(y, W, bounds, "tmp", init_weights, seed)
    param_names = [
        "A_HV",
        "A_HH",
        "B_HV",
        "B_HH",
        "C_HV",
        "C_HH",
        "alpha_HV",
        "alpha_HH",
        "gamma_HV",
        "gamma_HH",
        "D_HV",
        "D_HH",
    ] + ["S"]

    return y, y_hat, params, cost
```

```python
# estimate the biomass
def estimate_biomass(sar_list, 
                     agb_file, 
                     mask_file, 
                     out_name,
                     param0_file = None):
    
    '''
    estimate the biomass
    '''
    nwin_size = 0
    
    if not os.path.exists("output"):
        os.mkdir("output")
    out_prefix = f"output/{out_name}"
    mdl00 = mr.FieldRetrieval(mask_file, out_name=out_prefix)
    z1, W1, mask_ws1 = mdl00.inversion_setup(
        sar_list, agb_file=agb_file, mask_file=mask_file, n_1=nwin_size
    )
    z1_dim = z1.shape
    print(z1_dim)
    z1r = z1.reshape([-1, z1_dim[1] * z1_dim[2] * z1_dim[3]])
    z0 = mdl00.inversion_return_valid(z1r, mask_ws1, mask_file=mask_file)
    W0 = mdl00.inversion_return_valid(W1, mask_ws1, mask_file=mask_file)
    z0 = z0.reshape([-1, z1_dim[1], z1_dim[2], z1_dim[3]])

    import pickle

    with open(out_prefix + "_z0_W0.bin", "wb") as f:
        pickle.dump([z0, W0], f)

    if param0_file is None:
        param0_file = mdl00.params_calibration(z0, W0, out_prefix)

    W_mean, _ = mdl00.inversion_recursive_ws(z0, mask_file, param0_file=param0_file)

    y0 = W0
    m_t = W_mean.shape[1]
    y0 = y0.flatten()
    r2 = []
    rmse = []
    W_mean2 = []
    for i in range(m_t):
        # print(i)
        x0 = np.nanmean(W_mean[:, : i + 1], axis=1)
        r2_1 = r2_score(x0[~np.isnan(x0), None], y0[~np.isnan(x0), None])
        rmse_1 = np.sqrt(
            mean_squared_error(x0[~np.isnan(x0), None], y0[~np.isnan(x0), None])
        )
        W_mean2.append(x0)
        r2.append(r2_1)
        rmse.append(rmse_1)
        
    r2 = np.array(r2)
    rmse = np.array(rmse)
    
    return W0, W_mean2[-1], r2, rmse
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import gaussian_kde

sns.set()

def update_numbers_stats(y_true, y_pred):
    
    y = y_pred
    x = y_true
    
    xy = np.vstack([x, y])
    z0 = gaussian_kde(xy)(xy)
    
    p2 = np.sum((x - 20 < y) & (y < x + 20) & (x < 100)) / x[x < 100].size * 100
    p1 = np.sum((x - 10 < y) & (y < x + 10) & (x < 100)) / x[x < 100].size * 100
    
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    
    bias = np.mean(y) - np.mean(x)
    
    return x, y, z0, p1, p2, r2, rmse, bias


def plot(train_y_true, train_y_pred):
    
    '''
    plot the single prediction and estimation
    '''
    
    train_x, train_y, train_z0, train_p1, train_p2, train_r2, train_rmse, train_bias = update_numbers_stats(train_y_true, train_y_pred)

    plt.scatter(train_x, train_y, cmap = 'jet', c=train_z0, s=10, label="{:.1f}% within 20\n{:.1f}% within 10\nbias = {:.2f} Mg/ha\nrmse = {:.2f} Mg/ha".format(train_p2, train_p1, train_bias,train_rmse))
    plt.plot([0,300],[0,300])
    plt.xlabel('Measured AGB (Mg/ha)')
    plt.ylabel('Predicted AGB (Mg/ha)')
    plt.xlim([0,300])
    plt.ylim([0,300])
    plt.axes().set_aspect('equal', adjustable='box')
    plt.legend()
    
    plt.savefig('output/ml_train_plot.png',dpi=300, bbox_inches='tight', pad_inches=0)
    plt.show()
    
def plot_train_test(train_y_true, train_y_pred, test_y_true, test_y_pred):
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    
    train_x, train_y, train_z0, train_p1, train_p2, train_r2, train_rmse, train_bias = update_numbers_stats(train_y_true, train_y_pred)
    test_x, test_y, test_z0, test_p1, test_p2, test_r2, test_rmse,test_bias = update_numbers_stats(test_y_true, test_y_pred)

    ax1.scatter(train_x, train_y, cmap = 'jet', c=train_z0, s=10, label="{:.1f}% within 20\n{:.1f}% within 10\nbias = {:.2f} Mg/ha\nrmse = {:.2f} Mg/ha".format(train_p2, train_p1, train_bias,train_rmse))
    ax1.set_title('Train')
    ax1.plot([0,300],[0,300])
    ax1.set_xlabel('Measured AGB (Mg/ha)')
    ax1.set_ylabel('Predicted AGB (Mg/ha)')
    ax1.set_xlim([0,300])
    ax1.set_ylim([0,300])
    ax1.set_aspect('equal', adjustable='box')
    ax1.legend()
    
    ax2.scatter(test_x,  test_y,  cmap = 'jet', c=test_z0,  s=10, label="{:.1f}% within 20\n{:.1f}% within 10\nbias = {:.2f} Mg/ha\nrmse = {:.2f} Mg/ha".format(test_p2,  test_p1, test_bias, test_rmse))
    ax2.plot([0,300],[0,300])
    ax2.set_title('Test')
    ax2.set_xlabel('Measured AGB (Mg/ha)-Test')
    ax2.set_ylabel('Predicted AGB (Mg/ha)')
    ax2.set_xlim([0,300])
    ax2.set_ylim([0,300])
    ax2.set_aspect('equal', adjustable='box')
    ax2.legend()
    
    plt.savefig('output/ml_train_test_plot.png',dpi=300, bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
#randomly select about 1500 for each run

Y = []
Y_hat = []
Params = []
Costs = []

#iterate 10 times
number_of_iterations = 11
number_of_optimization = 100


for iterate in range(0,number_of_iterations):

    train_file, train_indices, train_indices_index = randomly_select_training_samples(calval_mask_file, 1500, None)

    test_file,  test_indices,  test_indices_index  = randomly_select_testing_samples(calval_mask_file, 
                                                                                     train_indices, 
                                                                                     train_indices_index, 
                                                                                     500, None)

    sar_reg_list = pd.read_csv("rs_data/uavsar/sar_reg_list_leno.csv")['file_name'].tolist()
    sar_reg_list = [f"rs_data/uavsar/{i}" for i in sar_reg_list]
    agb_file = "rs_data/lidar_biomass/agb_2019_reg.tif"

    #outlier and noise removal
    train_y0, train_xhv, train_xhh, train_clean_mask_file = update_numbers(sar_reg_list, agb_file, train_file, 'leno_train', 10, 10)
    test_y0, test_xhv, test_xhh, test_clean_mask_file = update_numbers(sar_reg_list, agb_file, test_file, 'leno_test', 10, 10)
    
    #save to the file
    with open(f"output/leno_train_test_set_{iterate}.bin", "wb") as f:
        pickle.dump([train_y0, train_xhv, train_xhh, test_y0, test_xhv, test_xhh], f)
        
    for k in range(0,number_of_optimization):
        #model calibration for each optimization
        y, y_hat, params, cost = model_calibration(train_y0, train_xhv, train_xhh, init_weights= [4,1], seed=None)

        #
        Y.append(y)
        Y_hat.append(y_hat)
        Params.append(params)
        Costs.append(cost)
```
